```python
import serial

baud_rate = 230400 #whatever baudrate you are listening to
com_port1 = '/dev/ttyUSB0' #replace with your com port path
com_port2 = '/dev/ttyUSB1' #foward

listener = serial.Serial(com_port1, baud_rate)
forwarder = serial.Serial(com_port2, baud_rate)
```

```python
import time

t0 = time.time()
t1 = 0

while t1 - t0 < 2:
    t1 = time.time()
print('finish')
```

```python
import time

t0 = time.time()
t1 = 0

while t1 - t0 < 5:
    serial_out = listener.read(size=1)
    print serial_out #or write it to a file 
    forwarder.write(serial_out)
    t1 = time.time()
print('done')
```

```python
listener.close();
forwarder.close();
```
